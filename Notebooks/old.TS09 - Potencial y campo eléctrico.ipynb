## Cálculo de campo eléctrico

Una vez que se conoce el potencial eléctico en una región, se puede obtener las componentes del campo eléctrico como

$$
E_x = -\frac{\partial V}{\partial x}    
$$

para cada eje. Donde el campo eléctrico $\vec E$
se expresa como

$$
\vec E = E_x\hat\imath + E_y \hat\jmath + E_z\hat k
$$

Se puede aplicar el método de diferencias finitas para estimar las componentes del campo:

$$
E_x(i,j) \approx -\frac{V(i+1,j)-V(i-1,j)}{2\Delta x}
$$
