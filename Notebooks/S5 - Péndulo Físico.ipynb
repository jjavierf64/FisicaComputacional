# Péndulo Físico

Para el caso final se va a considerar un oscilador más complicado: se considera la presencia de una fuerza externa y no se hace la aproximación para ángulos pequeños.

La fuerza externa tiene la forma de una función sinoidal con frecuencia $\Omega_D$. Si no se considera el amortiguamiento, esta frecuencia compite con la frecuencia $\omega_0$ del OAS.

El comportamiento más interesante del oscilador se da cuando se toman en cuenta los efectos del amortiguamiento, la no linealidad, y la fuerza externa:

La ecuación de movimiento se expresa como:
$$
\ddot \theta = -\frac{g}{l}\sin(\theta) - q \dot \theta + F_D\sin(\Omega_D t)
$$

Esta ecuación se expresa como el siguiente sistema de ecuaciones:
\begin{align}
\begin{cases}
    \dot \omega &= -\dfrac{g}{l}\sin(\theta) -q\omega + F_D\sin(\Omega_D t)\\
    \dot \theta &= \omega
\end{cases}
\end{align}
