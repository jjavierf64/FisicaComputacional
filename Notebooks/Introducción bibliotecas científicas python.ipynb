### Ejemplo: Rutina de integración

Definimos una función en el contexto de python para que nos retorne el valor de la función exponencial $$f(x) = e^{ax}$$para cierto valor de a:

```python
def FuncionExp(coord_X, a=1.):
    """
    Función que retorna el valor de la función e^(a*x).
    a: parámetro que se define por defecto como 1.
    """
    return np.exp(a*coord_X)
```

```python
x = 1
y = FuncionExp(x)ticks=10
print('Valor de la función exponencial evaluada en {}: {:.3f}'.format(x, y))
```

Ahora usamos la rutina de integración de SciPy para obtener:
$$\texttt{resultado_integral} = \int_{0}^{1}e^{ax}{\mathrm d}x$$

```python
from scipy import integrate

resultado_integral = integrate.quad(FuncionExp, 0, 1)
print('Resultado de la integral: {}'.format(resultado_integral))
```

### Ejemplo: Rutina de regresión

```python
import matplotlib.pyplot as plt

# Se genera una señal con ruido
nPuntos = 50
puntos_X = np.linspace(0,10,nPuntos)
señal_original = np.sin(puntos_X)
señal_ruidosa = np.sin(puntos_X)+np.random.normal(0,0.5,50)


print(puntos_X)
print(señal_ruidosa)

# gráfico de dispersión
fig, ax = plt.subplots(figsize=(7,7), dpi=100)
plt.plot(puntos_X, señal_ruidosa, marker='o', color='g', label='ruidosa')
plt.plot(puntos_X, señal_original, marker='o', color='c', label='original', alpha=0.2)


plt.title('Gráfico 1')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend(loc='best')
plt.show()
```

```python
from scipy import optimize

# Se define la función para el ajuste
def funcion_prueba(x, a, b):
    return a * np.sin(b * x)

parametros, parametros_covariancia = optimize.curve_fit(funcion_prueba, puntos_X, señal_ruidosa)
print(parametros)

señal_ajustada = np.zeros(nPuntos)
for i in range(nPuntos):
    señal_ajustada[i] = funcion_prueba(puntos_X[i], parametros[0], parametros[1])

# gráfico de dispersión
grafica, ejes = plt.subplots(figsize=(7,7), dpi=100)
ejes.plot(puntos_X, señal_ruidosa, marker='o', color='g', label='ruido', alpha=0.1)
ejes.plot(puntos_X, señal_ajustada, marker='o', color='m', label='ajuste')
ejes.plot(puntos_X, señal_original, marker='o', color='c', label='original', alpha=0.5)


plt.title('Gráfico 1')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend(loc='best')
plt.show()
```

```python
def lorenz(x, y, z, s=10, r=28, b=2.667):
    '''
    Given:
       x, y, z: a point of interest in three dimensional space
       s, r, b: parameters defining the lorenz attractor
    Returns:
       x_dot, y_dot, z_dot: values of the lorenz attractor's partial
           derivatives at the point x, y, z
    '''
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot


dt = 0.01
num_steps = 10000

# Need one more for the initial values
xs = np.empty(num_steps + 1)
ys = np.empty(num_steps + 1)
zs = np.empty(num_steps + 1)

# Set initial values
xs[0], ys[0], zs[0] = (0., 1., 1.05)

# Step through "time", calculating the partial derivatives at the current point
# and using them to estimate the next point
for i in range(num_steps):
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + (x_dot*dt)
    ys[i + 1] = ys[i] + (y_dot*dt)
    zs[i + 1] = zs[i] + (z_dot*dt)


# Plot
fig = plt.figure(dpi=120)
ax = fig.gca(projection='3d')

ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title('Atractor de Lorenz')

plt.show()
```
