1. Escriba una rutina llamada ```RandomWalk()``` que genere una caminata aleatoria de _nPasos_ pasos en _nDim_ dimensiones. La longitud de cada paso es aleatoria y est谩 distribuida uniformemente entre (-1/2, 1/2) en cada dimensi贸n.
2. Genere los pasos como un arreglo de valores aleatorios de _nPasos x nDim_ usando `np.random.uniform()`, luego use la funci贸n `np.cumsum()` para hacer una suma acumulativa. Esto resulta en un arreglo con la posici贸n en la caminata luego de cada paso. **Valor: 2 puntos.**

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Base para la funci贸n RandomWalk()

def RandomWalk(nPasos, nDim):
    """
    Use random.uniform(min, max, shape) to generate an array of steps of shape (N,d), 
    and then use np.cumsum(..., axis=0) (which adds them up along the 'N' axis).
    """
    steps = np.random.uniform(-1/2, 1/2, (nPasos,nDim))
    walks = np.cumsum(steps, axis=0)
    return walks
````

5. Cada caminata aleatoria es diferente e impredecible pero el _ensamble_ de caminatas tiene propiedades elegantes y predecibles. Escriba una rutina llamada ```EndPoints()``` que calcule la posici贸n final de _nCaminatas_ caminatas aleatorias de _nPasos_ pasos en _nDim_ dimensiones. **Valor: 1 punto.**

````python
# Base para la funci贸n Endpoints()

def Endpoints(nCaminatas, nPasos, nDim):
    
    puntosFinales = []
    
    for i in range(nCaminatas):
        steps = np.random.uniform(-1/2, 1/2, (nPasos,nDim))
        walks = np.cumsum(steps, axis=0)
        puntosFinales.append(walks[-1])
        
    return puntosFinales
````

9. Escriba una rutina que grafique un histograma de los puntos finales de _nCaminatas = 10 000_ en una dimensi贸n. Genere histogramas para _nPasos = 1, 2, 3, 5, y 10_ utilizando 50 _bins_.
10. Sobre el mismo histograma, grafique la predicci贸n dada por la distribuci贸n gaussiana $\rho(x)$ para _x_ entre $(-3\sigma, 3\sigma)$. **Valor: 2 puntos.**

````python

def Endpoints1D(nCaminatas, nPasos):
    
    steps = np.random.uniform(-1/2, 1/2, (nCaminatas,nPasos))
    walks = np.cumsum(steps, axis=1)
    puntosFinales = walks[:,-1]
        
    return puntosFinales
````

````python
# Histograma

# Definici贸n de la funci贸n de distribuci贸n Gaussiana
def distrGauss(t, sigma):
    return (1/(np.sqrt(2 * np.pi) * sigma)) * np.exp(-(t**2)/(2 * sigma**2))


nCaminatas = 10000
nPasosAll = (1,2,3,5,10)


for nPasos in nPasosAll:
    # Puntos finales de las caminatas
    puntos = Endpoints1D(nCaminatas, nPasos)
    
    R_rms = np.std(puntos)
    
    
    # Distribuci贸n Gaussiana
    x = np.arange(-3*R_rms, 3*R_rms, 0.1*R_rms)
    y = distrGauss(x,R_rms)
    
    
    
    # Graficos
    fig, ax = plt.subplots()
    
    ax.hist(puntos, bins=50, density=True, label="Puntos Finales")
    ax.plot(x,y, label="Distribuci贸n Gausssiana")
    
    ax.set_title("Histograma de puntos finales para "+str(nPasos)+" pasos.")
    ax.set_xlabel("Posici贸n Final")
    ax.set_ylabel("Frecuencia")
    ax.legend(loc="best")
    
    plt.show()



````
